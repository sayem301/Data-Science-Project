```python
import pandas as pd
import matplotlib.pyplot as plt


file_path="land_plot.xlsx"
data = pd.read_excel(file_path)

year = data['Year']
total_land = data['Total Land (Squ.km)']
number_of_char = data['Number of Char']

fig, ax1 = plt.subplots(figsize=(10, 6))

ax1.plot(year, total_land, color='Red', label='Total Land (Squ.km)')
ax1.set_xlabel('Year')
ax1.set_ylabel('Total Land (Squ.km)', color='Red')
ax1.tick_params(axis='y', labelcolor='Red')
ax1.set_ylim(30)  # Start the y-axis from 30
ax1.set_xlim(1970, year.max())  # Start the x-axis from 1970

ax2 = ax1.twinx()
ax2.plot(year, number_of_char, color='green', label='Number of Char')
ax2.set_ylabel('Number of Char', color='green')
ax2.tick_params(axis='y', labelcolor='green')
ax2.set_ylim(0)  # Start the y-axis from 1


ax1.grid(True, which='major', linestyle='--', linewidth=0.3)


lines_1, labels_1 = ax1.get_legend_handles_labels()
lines_2, labels_2 = ax2.get_legend_handles_labels()
ax1.legend(lines_1 + lines_2, labels_1 + labels_2, loc='upper left', bbox_to_anchor=(0.05, 1), ncol=1,)


plt.tight_layout()
plt.show()
#plt.savefig('F:/Education/Geology/higher study/kukri_mukri/kukri_mukri/table_data/python/land.png', format='png', dpi=300, bbox_inches='tight')
plt.close()

print("Plot saved successfully.")
```
