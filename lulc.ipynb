```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

data = pd.read_excel('all.xlsx')

years = data.columns[1:].values  
forest = data.iloc[0, 1:].values  

X = years.reshape(-1, 1) 
y = forest  

model = LinearRegression()

model.fit(X, y)

prediction_years = np.arange(1973, 2041).reshape(-1, 1)
predicted_forest_values = model.predict(prediction_years)

plt.figure(figsize=(10, 6))
plt.plot(prediction_years, predicted_forest_values, color='red', label='Predicted Forest', linewidth=2)

actual_years = [1973, 1980, 1990, 2000, 2010, 2020]
actual_values = forest  
plt.scatter(actual_years, actual_values, color='blue', label='Actual Forest', zorder=5)

plt.xlabel('Year')
plt.ylabel('Forest Area (km²)')

plt.legend(
    loc='upper left',        
    fontsize=13,             
    #bbox_to_anchor=(1.05, 1),
    #frameon=True,            
    borderpad=1,            
    labelspacing=0.5        
)

#plt.grid(True)
plt.tight_layout()
plt.show()
#plt.savefig('F:/Education/Geology/higher study/kukri_mukri/kukri_mukri/table_data/python/forrest.png', format='png', dpi=300, bbox_inches='tight')
plt.close()

print("Plot saved successfully.")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

data = pd.read_excel('all.xlsx')

years = data.columns[1:].values  
cropland = data.iloc[1, 1:].values  


X = years.reshape(-1, 1)  
y = cropland  


model = LinearRegression()


model.fit(X, y)


prediction_years = np.arange(1973, 2041).reshape(-1, 1)
predicted_cropland_values = model.predict(prediction_years)


plt.figure(figsize=(10, 6))
plt.plot(prediction_years, predicted_cropland_values, color='green', label='Predicted Cropland', linewidth=2)


actual_years = [1973, 1980, 1990, 2000, 2010, 2020]
actual_values = cropland  # Actual cropland values for these years
plt.scatter(actual_years, actual_values, color='blue', label='Actual Cropland', zorder=5)


plt.xlabel('Year')
plt.ylabel('Cropland Area (km²)')
plt.title('Predicted Cropland Values (1973-2030) with Actual Points')

plt.legend(
    loc='upper left',        
    fontsize=13,             
    borderpad=1,             
    labelspacing=0.5        
)


plt.tight_layout()

plt.show()
#plt.savefig('F:/Education/Geology/higher study/kukri_mukri/kukri_mukri/table_data/python/cropland.png', format='png', dpi=300, bbox_inches='tight')
plt.close()

print("Plot saved successfully.")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression


data = pd.read_excel('all.xlsx')


years = data.columns[1:].values  
wetland = data.iloc[2, 1:].values  


X = years.reshape(-1, 1)  
y = wetland 


model = LinearRegression()


model.fit(X, y)


prediction_years = np.arange(1973, 2041).reshape(-1, 1)
predicted_wetland_values = model.predict(prediction_years)


plt.figure(figsize=(10, 6))
plt.plot(prediction_years, predicted_wetland_values, color='purple', label='Predicted Wetland', linewidth=2)


actual_years = [1973, 1980, 1990, 2000, 2010, 2020]
actual_values = wetland  
plt.scatter(actual_years, actual_values, color='blue', label='Actual Wetland', zorder=5)


plt.xlabel('Year')
plt.ylabel('Wetland Area (km²)')
plt.title('Predicted Wetland Values (1973-2030) with Actual Points')

plt.legend(
    loc='upper left',        
    fontsize=13,             
    borderpad=1,             
    labelspacing=0.5         
)


plt.tight_layout()

plt.show()
#plt.savefig('F:/Education/Geology/higher study/kukri_mukri/kukri_mukri/table_data/python/wetland.png', format='png', dpi=300, bbox_inches='tight')
plt.close()

print("Plot saved successfully.")
```
